```python
import numpy as np
import matplotlib.pyplot as plt
from numpy import pi

a = np.arange(10, 30, 5) # 0 30 5
b = np.arange(0, 2, 0.3) # 0 2 0.3

print(a)
print(b)

c = np.linspace(0, 2, 15) #0 ~ 2 15 

print(c)

x = np.linspace(0, 2*pi, 100)
f = np.sin(x)
print(f)

points = np.arange(-5, 5, 0.01)
xs, ys = np.meshgrid(points, points)

z = np.sqrt(xs**2 + ys**2)

plt.imshow(z, cmap=plt.cm.gray)
plt.colorbar()
plt.title("image")
```
