## 10장. 자동차연비분석

```python
import warnings

warnings.filterwarnings(action='ignore') 
```

```python
import numpy as np
import pandas as pd

data_df = pd.read_csv('auto-mpg.csv', header=0, engine='python')
```

# 2) 데이터 준비 및 탐색

```python
print(' 데이터셋 크기 : ', data_df.shape)

data_df.head()
```

```python
data_df = data_df.drop(['car_name', 'origin', 'horsepower'], axis=1, inplace=False)
```

```python
print('데이터세트 크기 : ',data_df.shape)

data_df.head()
```

```python
data_df.info()
```

## 3) 분석 모델 구축

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
```

```python
# X,Y 분할하기

Y = data_df['mpg']
X = data_df.drop(['mpg'], axis=1, inplace=False)
```

```python
# 훈련용 데이터와 평가용 데이터 분할하기
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=0)
```

```python
# 선형회귀분석 : 모델 생성
lr = LinearRegression()
```

```python
# 선형회귀분석 : 모델 훈련
lr.fit(X_train, Y_train)
```

```python
# 선형회귀분석 : 평가 데이터에 대한 예측 수행 -> 예측 결과 Y_predict 구하기
Y_predict = lr.predict(X_test)
```

## 4) 결과 분석 및 시각화

```python
mse = mean_squared_error(Y_test,Y_predict)
rmse = np.sqrt(mse)

print('MSE : {0:.3f}, RMSE : {1:.3f}'.format(mse,rmse))
print('R^2(Variance score) : {0:.3f}'.format(r2_score(Y_test,Y_predict)))
```

```python
print('Y 절편 값: ',  np.round(lr.intercept_, 2))
print('회귀 계수 값: ', np.round(lr.coef_, 2))
```

```python
coef = pd.Series(data=np.round(lr.coef_, 2), index=X.columns)
coef.sort_values(ascending=False)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
fig, axs = plt.subplots(figsize=(16, 16), ncols=3, nrows=2)

x_features = ['model_year', 'acceleration', 'displacement', 'weight', 'cylinders']
plot_color = ['r', 'b', 'y', 'g', 'r']

for i, feature in enumerate(x_features):
      row = int(i/3)
      col = i%3
      sns.regplot(x=feature, y='mpg', data=data_df, ax=axs[row][col], color=plot_color[i])
```

### 1) 데이터 수집

```python
import numpy as np
import pandas as pd

from sklearn.datasets import fetch_california_housing
california = fetch_california_housing()
```

### 2) 데이터 준비 및 탐색

```python
print(california.DESCR)
```

```python
california_df = pd.DataFrame(california.data, columns = california.feature_names)
california_df.head()
```

```python
california_df['PRICE'] = california.target
california_df.head()
# MedInc(Median Income) : 해당 지역 가구의 중간 소득
# HouseAge(Median House Age) : 해당 지역 주택의 중간 연식
# AveRooms(Average Rooms) : 해당 지역 주택의 평균 방 수
# Population : 해당 지역 인구수
# AveOccup(Average Occupancy) : 해당 지역 주택의 평균 입주율
# Latitude : 해당지역 위도
# Longitude : 해당 지역 경도
```

```python
print('캘리포니아 주택 가격 데이터셋 크기 : ', california_df.shape)
```

```python
california_df.info()
```

### 3) 분석 모델 구축

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
```

```python
# X, Y 분할하기
Y = california_df['PRICE']
X = california_df.drop(['PRICE'], axis=1, inplace=False)
```

```python
# 훈련용 데이터와 평가용 데이터 분할하기
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.3, random_state=156)
```

```python
# 선형회귀분석 : 모델 생성
lr = LinearRegression()
```

```python
# 선형회귀분석 : 모델 훈련
lr.fit(X_train, Y_train)
```

```python
# 선형회귀분석 : 평가 데이터에 대한 예측 수행 -> 예측 결과 Y_predict 구하기
Y_predict = lr.predict(X_test)
```

### 4) 결과 분석 및 시각화

```python
mse = mean_squared_error(Y_test, Y_predict)
rmse = np.sqrt(mse)

# train 이 얼마나 잘 됐는지를 파악하기 위해 보는 자료
print('MSE : {0:.3f}, RMSE : {1:.3f}'.format(mse,rmse))
print('R^2(Variance score) : {0:.3f}'.format(r2_score(Y_test, Y_predict)))
```

```python
print('Y 절편 값: ', lr.intercept_)
print('회귀 계수 값: ', np.round(lr.coef_, 1))
```

```python
coef = pd.Series(data = np.round(lr.coef_, 2), index = X.columns)
coef.sort_values(ascending = False)
```

```python
print(b_cancer.data[0])
```

```python
print(b_cancer_scaled[0])
```

## 2) 분석 모델 구축 : 로지스틱 회귀를 이용한 이진 분석 모델

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
# X, Y 설정하기
Y = b_cancer_df['diagnosis']
X = b_cancer_scaled
```

```python
# 훈련용 데이터와 평가용 데이터 분할하기
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.3, random_state=0)
```

```python
# 로지스틱 회귀 분석 : (1) 모델 생성
lr_b_cancer = LogisticRegression()
```

```python
# 로지스틱 회귀 분석 : (2) 모델 훈련
lr_b_cancer.fit(X_train, Y_train)
```

```python
# 로지스틱 회귀 분석 : (3) 평가 데이터에 대한 예측 수행 -> 예측 결과 Y_predict 구하기
Y_predict = lr_b_cancer.predict(X_test)
```

## 4) 결과 분석

```python
from sklearn.metrics import confusion_matrix, accuracy_score
from sklearn.metrics import precision_score, recall_score, f1_score, roc_auc_score
```

```python
# 오차 행렬
confusion_matrix(Y_test, Y_predict)
```

```python
accuracy = accuracy_score(Y_test, Y_predict)
precision = precision_score(Y_test, Y_predict)
recall = recall_score(Y_test, Y_predict)
f1 = f1_score(Y_test, Y_predict)
roc_auc = roc_auc_score(Y_test, Y_predict)
```

```python
print('정확도 : {0:.3f}, 정밀도: {1:.3f}, 재현율: {2:.3f}, F1: {3:.3f}'.format(accuracy,precision,recall,f1))
```

```python
print('ROC_AUC: {0:.3f}'.format(roc_auc))
```
